```python
result = marketAPI.get_tickers('SWAP')

Result=result['data']
len(Result) 
```

```python
i=0
R=[]

while i < len(Result):
    instrument_id = Result[i]['instId']
    R.append(instrument_id)
    i=i+1    
```

```python
yoy=[]
for XXX in R:
    result = publicAPI.funding_rate_history(instId=XXX,limit='100')['data']
    kok = pd.DataFrame(result,columns=['instId','instType','fundingRate','realizedRate','fundingTime'])
    yoy.append(kok)
    time.sleep(0.1)
yoy_result=pd.concat(yoy)
```

```python
pd.set_option('display.max_rows',None) 
```

```python
y=[]
for x in yoy_result['instId'].unique():
    if x!='instId':
        y.append(x)
youmean=[]
you=[]
you1mean=[]
you1=[]
you2mea=[]
you2=[]
you2mean=[]
for x in y:
    qoo=yoy_result[yoy_result['instId']==x]
    qoo.sort_values(by="fundingTime",ascending=False)
    #print(qoo)
    qoo['realizedRate']=pd.to_numeric(qoo['realizedRate'])
    bb=qoo['realizedRate'].values
    youmean.append(bb.mean()*100)
    you.append(bb.mean()/bb.std(ddof=1))
    you1mean.append(bb[:3].mean()*100)
    you1.append(bb[:3].mean()/bb[:3].std(ddof=1))
    you2mean.append(bb[:30].mean()*100)
    you2.append(bb[:30].mean()/bb[:30].std(ddof=1))
    data={'Mean':youmean,'Allindex':you,'lastnightMean':you1mean,'lastnight':you1,'last10daysMean':you2mean,'last10days':you2}
youou=pd.DataFrame(data,index=y,columns=['Mean','Allindex','lastnightMean','lastnight','last10daysMean','last10days'])
youou.sort_values(by=['last10daysMean'],ascending=False)
```
